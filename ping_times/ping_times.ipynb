## ETL: Parse the data and save to CSV

```python
f = open("ping.log", 'r')
g = open("ping.csv", 'w')

for row in f:
    t = row.split(" ")
    if len(t) < 10: continue
    datestamp = t[:2]
    datestring = " ".join(datestamp)
    datestring=datestring.replace(',', '.')
    
    temp = t[-2]
    u = temp.split("=")
    if len(u) < 2:
        continue
    value = u[1]
    g.write("%s,%s\n" % (datestring, value))
    
g.close()
f.close()
```
